```python
# Function call API
model = "glm-3-turbo"
tools =  [
{   'type' :'function',
   "function" : 
      {
        'name':'get_connections',
        'description': '判断一句话中的两个用户1和用户2之间',
        "parameters": {
        "type": "object",
            "properties": {
              "user_id1": {
                "type": "string",
                "description": "用户1 名称"
              },
              "user_id2": {
                "type": "string",
                "description": "用户2 名称"
              },
            },
        "required": ["user1_name", "user2_name"]
        }   
      },   
    
}
]
response = client.chat.completions.create(
    model = model,  # 填写需要调用的模型名称
    messages=[
        {"role": "user", "content": "李华和小王是不是认识？"},
    ],
    tools = tools,
)
print(response,'\n',response.choices[0].message)
```

```python
# Function call API
model = "glm-3-turbo"
tools =  [
{   'type' :'function',
   "function" : 
      {
        'name':'get_connections',
        'description': '判断句子中车票的信息，包括始发地，目的地以及出发时间等',
        "parameters": {
        "type": "object",
            "properties": {
              "departure": {
                "type": "string",
                "description": "始发地"
              },
              "destination": {
                "type": "string",
                "description": "目的地"
              },
              "start_time":{
                  'type':'string',
                  'description':'出发时间'
              },
                # 两个有一个就行
              "end_time":{
                  'type':'string',
                  'description':'到达时间'
              }
            },
        "required": ["departure", "destination","start_time","end_time",]  # 两个有一个就行
        }   
      },   
    
}
]
response = client.chat.completions.create(
    model = model,  # 填写需要调用的模型名称
    messages=[
        {"role": "user", "content": "你能帮我查询2024年1月1日12点30分从北京南站到上海的火车票吗？大概半小时的车程"},
    ],
    tools = tools,
)
print(response,'\n',response.choices[0].message)
```
